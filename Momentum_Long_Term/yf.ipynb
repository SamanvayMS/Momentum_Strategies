```python
import yfinance as yf

msft = yf.Ticker("MSFT")

# Get all stock info
# info = msft.info
# print(info)

# Get historical market data
hist = msft.history(period="1mo")
print(hist)
```

```python
# Show meta information about the history
meta_info = msft.history_metadata
print(meta_info)
```

```python

# Show actions (dividends, splits)
actions = msft.actions
dividends = msft.dividends
splits = msft.splits
print(actions)
print(dividends)
print(splits)

# Note: `msft.capital_gains` is not a method provided by yfinance.
```

```python
# Show share count
share_count = msft.get_shares_full(start="2022-01-01", end=None)
print(share_count)
```

```python
# Show financials:
# - Income Statement
income_statement = msft.income_stmt
quarterly_income_statement = msft.quarterly_income_stmt
print(income_statement)
print(quarterly_income_statement)
```

```python

# - Balance Sheet
balance_sheet = msft.balance_sheet
quarterly_balance_sheet = msft.quarterly_balance_sheet
print(balance_sheet)
print(quarterly_balance_sheet)
```

```python

# - Cash Flow Statement
cashflow = msft.cashflow
quarterly_cashflow = msft.quarterly_cashflow
print(cashflow)
print(quarterly_cashflow)
```

```python

# Show holders
major_holders = msft.major_holders
institutional_holders = msft.institutional_holders
mutualfund_holders = msft.mutualfund_holders
#insider_transactions = msft.insider_transactions
#insider_purchases = msft.insider_purchases
#insider_roster_holders = msft.insider_roster_holders

print(major_holders)
print(institutional_holders)
print(mutualfund_holders)
#print(insider_transactions)
#print(insider_purchases)
#print(insider_roster_holders)
```

```python

# Show recommendations
# recommendations = msft.recommendations
# recommendations_summary = msft.recommendations_summary
# upgrades_downgrades = msft.upgrades_downgrades
# print(recommendations)
# print(recommendations_summary)
# print(upgrades_downgrades)

# Show future and historic earnings dates
earnings_dates = msft.earnings_dates
print(earnings_dates)
```

```python

# Show ISIN code - Experimental feature
# isin = msft.isin
# print(isin)

# Show options expirations
options = msft.options
print(options)
```

```python
# Show news
news = msft.news
for article in news:
    for key in article.keys():
        print(key, article[key])
```

```python

# Get option chain for specific expiration
# Replace 'YYYY-MM-DD' with the desired date
opt = msft.option_chain('2024-01-12')
calls = opt.calls
puts = opt.puts
print(calls)
print(puts)
```
